```python
with open("config.txt", "w") as cfg:
    cfg.write("db_url=oldhost:3306\nusername=admin\npassword=oldpass")


config = {}
with open("config.txt", "r") as cfg:
    for line in cfg:
        if '=' in line:
            key, value = line.strip().split('=', 1)
            config[key] = value

print("Original Config:", config)


config["db_url"] = "newhost:5432"
config["password"] = "newpass123"


with open("config.txt", "w") as cfg:
    for key, value in config.items():
        cfg.write(f"{key}={value}\n")


with open("config.txt", "r") as cfg:
    updated_config = cfg.read()
    print("\nUpdated Config File Content:\n", updated_config)
```
