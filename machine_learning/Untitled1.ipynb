```python
def id():
    name = input("Enter your name: ")
    age = input("Enter your age: ")
    branch = input("Enter your branch: ")
    print(f"Name: {name}, Age: {age}, Branch: {branch}")
```

```python
id()
```

```python
my_integer = 10
my_float = 3.14
my_string = "lakshya"

print("\nMemory Addresses:")
print(f"1st var: {id(my_integer)}")
print(f"2nd var: {id(my_float)}")
print(f"3rd var: {id(my_string)}")
```

```python
print("\nMemory Addresses:")
print(f"1st var: {type(my_integer)}")
print(f"2nd var: {type(my_float)}")
print(f"3rd var: {type(my_string)}")
```
