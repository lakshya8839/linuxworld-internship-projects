```python
import os
from base58 import b58encode
from solana.keypair import Keypair

# Goat SDK imports
from goat_wallets.solana import solana, send_solana
from goat_wallets.solana.client import SolanaClient
from goat_wallets.solana.utils import Keypair as GoatKeypair
from goat_adapters.langchain import get_on_chain_tools
from goat_plugins.spl_token import spl_token, SplTokenPluginOptions
from goat_plugins.spl_token.tokens import SPL_TOKENS

# ✅ STEP 1: Generate a fresh Solana keypair (safe for devnet)
keypair = Keypair.generate()
public_key = str(keypair.public_key)
private_key_base58 = b58encode(keypair.seed).decode()

print("✅ Generated Wallet")
print("📬 Public Address:", public_key)
print("🔐 Private Key (base58):", private_key_base58)

# ✅ STEP 2: Set up Solana RPC client (DEVNET)
rpc_url = "https://api.devnet.solana.com"
client = SolanaClient(rpc_url)

# ✅ STEP 3: Create wallet from private key
goat_keypair = GoatKeypair.from_base58_string(private_key_base58)
wallet = solana(client, goat_keypair)

# ✅ STEP 4: Set up plugins and tools
tools = get_on_chain_tools(
    wallet=wallet,
    plugins=[
        send_solana(),
        spl_token(SplTokenPluginOptions(
            network="devnet",  # Use devnet for testing!
            tokens=SPL_TOKENS
        )),
    ],
)

# ✅ STEP 5: Use tool to get balance
balance = tools["get_balance"].invoke({ "address": public_key })
print("💰 SOL Balance:", balance)
```
