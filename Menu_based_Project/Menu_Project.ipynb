```python
import os

print(
    """
    ---------------------------------------------------
    Linux Menu Based Project - Run any command of Linux
    ---------------------------------------------------
    1. Date Command
    2. Cal command
    3. Ifconfig command
    4. ls command
    5. mkdir command
    6. useradd command
    7. userdel command
    8. list all file permissions
    9. show particular file permissions
    10.Change file permissions
    11. Exit
    """
)

user = input("Enter username: ")
ip = input("Enter your remote IP: ")

while True:
    choice = input("Enter your choice: ")
    if choice == "1":
        os.system(f"ssh {user}@{ip} date")
    elif choice == "2":
        os.system(f"ssh {user}@{ip} cal")
    elif choice == "3":
        os.system(f"ssh {user}@{ip} ifconfig")
    elif choice == "4":
        os.system(f"ssh {user}@{ip} ls")
    elif choice == "5":
        new_directory_name = input("Enter new directory name: ")
        os.system(f"ssh {user}@{ip} mkdir {new_directory_name}")
    elif choice == "6":
        useradd = input("Enter username to add: ")
        os.system(f"ssh {user}@{ip} useradd {useradd}")
    elif choice == "7":
        userdel = input("Enter username to delete: ")
        os.system(f"ssh {user}@{ip} userdel {userdel}")
    elif choice == "8":
        os.system(f"ssh {user}@{ip} ls -l")
    elif choice == "9":
        file_name = input("File name whose permissions to be checked? : ")
        os.system(f"ssh {user}@{ip} ls -l {file_name}")
    elif choice == "10":
        change_file_permissions = input("Tell the permission you want to give/remove from user: ")
        option = input("Want to add(+) / want to remove(-) permissions? = ")
        type_of_access = input("User(u) / Group(g) / Other(o) to whom you want to give/remove permissions: ")
        file_name = input("File name whose permissions to be edited? : ")
        os.system(f"ssh {user}@{ip} chmod {type_of_access}{option}{change_file_permissions} {file_name}")
    elif choice == "11":
        break
    else:
        print("Invalid choice. Try again.")
```
