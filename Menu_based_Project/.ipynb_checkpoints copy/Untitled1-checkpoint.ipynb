```python
import os
from base58 import b58encode
from solana.keypair import Keypair
from goat_wallets.solana import solana
from goat_adapters.langchain import get_on_chain_tools
from goat_plugins.spl_token import spl_token, SplTokenPluginOptions
from goat_plugins.spl_token.tokens import SPL_TOKENS

# ✅ Step 1: Generate new wallet
keypair = Keypair.generate()
public_key = str(keypair.public_key)
private_key_base58 = b58encode(keypair.seed).decode()

print("🔐 Generated Solana Wallet")
print("📬 Public Address:", public_key)
print("🧬 Base58 Private Key:", private_key_base58)

# ✅ Step 2: Setup RPC & environment
os.environ["SOLANA_RPC_ENDPOINT"] = "https://api.devnet.solana.com"
os.environ["SOLANA_WALLET_SEED"] = private_key_base58

# ✅ Step 3: Initialize wallet using goat_wallets.solana
wallet = solana(
    os.environ["SOLANA_RPC_ENDPOINT"],
    keypair
)

# ✅ Step 4: Initialize on-chain tools
tools = get_on_chain_tools(
    wallet=wallet,
    plugins=[
        spl_token(SplTokenPluginOptions(
            network="devnet",  # or "mainnet"
            tokens=SPL_TOKENS
        )),
    ],
)

# ✅ Step 5: Invoke balance check
result = tools["get_balance"].invoke({
    "address": public_key
})

print("💰 Wallet Balance:", result)
```

```python
import os
from goat_adapters.langchain import get_on_chain_tools
from goat_plugins.spl_token import spl_token, SplTokenPluginOptions
from goat_plugins.spl_token.tokens import SPL_TOKENS
from goat_wallets.solana import solana  # ✅ available
from solana.keypair import Keypair      # ✅ from solana-py (not goat_wallets)

# Step 1: Set or generate wallet
wallet_seed = os.getenv("SOLANA_WALLET_SEED")

if wallet_seed:
    keypair = Keypair.from_base58_string(wallet_seed)
else:
    keypair = Keypair.generate()
    print("🔐 New wallet base58 seed:", keypair.to_base58())

# Step 2: Create wallet using goat_wallets.solana.solana()
wallet = solana("https://api.devnet.solana.com", keypair)  # ✅ WORKS with your structure

# Step 3: Setup tools
tools = get_on_chain_tools(
    wallet=wallet,
    plugins=[
        spl_token(SplTokenPluginOptions(
            network="devnet",
            tokens=SPL_TOKENS
        )),
    ],
)

# Step 4: Get address & check balance
address = wallet.get_address()  # or wallet.address or wallet.public_key(), if get_address() doesn't exist
print("📬 Wallet address:", address)

result = tools["get_balance"].invoke({"address": address})
print("💰 Balance:", result)
```
