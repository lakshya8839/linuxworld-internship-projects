```python
from pymongo import MongoClient
```

```python
myclient = MongoClient( host = "127.0.0.1", port = 27017 )
```

```python
mydb = myclient["lwindia"]
```

```python
mycollection = mydb["student"]
```

```python
mydb.mycollection.insert_one(
    {
        "name" : "lakshya",
        "phoneno" : 1111,
        "course" : "d1"
    }
)
```

```python
mydb.mycollection.find_one()
```
